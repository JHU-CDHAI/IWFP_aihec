### Lifetime Drug Use


```python

recode_vars = ['a53_a54', 'a54', 'a55', 'a56', 'a57', 'a58', 'a59', 'a60', 'a61', 'a62', 'a63_a56']
recode_dict = {1: 0, 2: 1, 3: 2, 4: 3, 5: 4, 6: 5}

for var in recode_vars:
    df[var] = df[var].replace(recode_dict)


df['lifedrug'] = df[recode_vars].sum(axis=1) / len(recode_vars)
df['lifedrug'].value_counts()

scaler = StandardScaler()
df['zlifedrug'] = scaler.fit_transform(df[['lifedrug']])


df = df.dropna(subset=['zlifedrug'])
df['zlifedrug'].value_counts()
```

### Drug Education 


```python
# Define paired lists
orig_vars = ['a87_a78', 'a88_a79', 'a89_a80', 'a90_a81', 'a91_a82', 'a92_a83', 'a93_a84', 'a94_a85']
rev_vars = ['rev_a87_a78', 'rev_a88_a79', 'rev_a89_a80', 'rev_a90_a81',
            'rev_a91_a82', 'rev_a92_a83', 'rev_a93_a84', 'rev_a94_a85']
# Loop to generate reverse-coded versions
for orig, rev in zip(orig_vars, rev_vars):
    df[rev] = df[orig].replace({1: 4, 2: 3, 3: 2, 4: 1})

df['druged'] = df[rev_vars].mean(axis=1)
scaler = StandardScaler()
df['zdruged'] = scaler.fit_transform(df[['druged']])

df = df.dropna(subset=['zdruged'])


#  Compute composite and standardized score
df['druged'] = (df['rev_a87_a78'] + df['rev_a88_a79'] + df['rev_a89_a80'] + df['rev_a90_a81'] +
                df['rev_a91_a82'] + df['rev_a92_a83'] + df['rev_a93_a84'] + df['rev_a94_a85']) / 8


scaler = StandardScaler()
df['zdruged'] = scaler.fit_transform(df[['druged']])


df = df.dropna(subset=['zdruged'])
df['zdruged'].value_counts()
```

# Model


```python

X = df[['gender', 'trans', 'lgbt', 'foster', 'depression', 'homeless', 'breakfast',
        'zparented', 'zacademic', 'zlifealc', 'zlifecig', 'zlifevape', 'zlifecan', 
        'zlifedrug', 'zdruged', 'zdrugaccess', 'zbully']]
y = df['si']
```

```python
X
```

```python
from sklearn.linear_model import LogisticRegressionCV
model = LogisticRegressionCV(penalty='l1', solver='liblinear', cv=10, random_state=1533).fit(X, y)
```

```python
coef_standardized = model.coef_ / StandardScaler().fit(X).scale_
print("Standardized Coefficients:")
for name, coef in sorted(zip(X.columns, coef_standardized[0]), key=lambda x: abs(x[1]), reverse=True):
    print(f"{name}: {coef}")
```

# Predcitions and Performance

```python
from sklearn.metrics import roc_auc_score, brier_score_loss, precision_recall_curve
from sklearn.calibration import calibration_curve

si_estimatesPR3 = model.predict_proba(X)[:, 1]

auc = roc_auc_score(y, si_estimatesPR3)
print(f"AUC: {auc:.2f}")


prob_true, prob_pred = calibration_curve(y, si_estimatesPR3, n_bins=10)

import matplotlib.pyplot as plt
plt.plot(prob_pred, prob_true, 's-')
plt.plot([0, 1], [0, 1], '--', color='gray')
plt.xlabel('Predicted probability')
plt.ylabel('Actual probability')
plt.title('Calibration curve')
plt.show()

# Brier
brier = brier_score_loss(y, si_estimatesPR3)
print(f"Brier score: {brier:.2f}")


import numpy as np
print("Predicted probability distribution:")
print(f"Minimum: {np.min(si_estimatesPR3):.2f}")
print(f"25th percentile: {np.percentile(si_estimatesPR3, 25):.2f}") 
print(f"Median: {np.median(si_estimatesPR3):.2f}")
print(f"75th percentile: {np.percentile(si_estimatesPR3, 75):.2f}")
print(f"Maximum: {np.max(si_estimatesPR3):.2f}")
```
